```python
# Install libraries
!pip install -q pandas scikit-learn streamlit joblib

import pandas as pd
import joblib
from sklearn.ensemble import RandomForestRegressor
from google.colab import files

# =========================
# Create Dataset
# =========================

data = {
    "Project_Size":[100,150,200,250,300,350,400],
    "Team_Size":[5,6,8,10,12,15,18],
    "Duration_Months":[3,4,5,6,8,10,12],
    "Budget":[50000,70000,90000,120000,150000,180000,220000]
}

df = pd.DataFrame(data)

df.to_csv("budget_dataset.csv",index=False)

# =========================
# Train Model
# =========================

X = df[["Project_Size","Team_Size","Duration_Months"]]
y = df["Budget"]

model = RandomForestRegressor(random_state=42)
model.fit(X,y)

joblib.dump(model,"model.pkl")

# =========================
# Create app.py
# =========================

app_code = '''
import streamlit as st
import pandas as pd
import joblib

model = joblib.load("model.pkl")

st.title("Project Budget Prediction Web App")

st.write("Enter project details below:")

size = st.number_input("Project Size", min_value=1)

team = st.number_input("Team Size", min_value=1)

duration = st.number_input("Duration (Months)", min_value=1)

if st.button("Predict Budget"):

    input_data = pd.DataFrame(
        [[size,team,duration]],
        columns=[
            "Project_Size",
            "Team_Size",
            "Duration_Months"
        ]
    )

    prediction = model.predict(input_data)

    st.success(
        f"Estimated Budget: ${prediction[0]:,.2f}"
    )

st.subheader("Model Explanation")

st.write(
"This application uses a Random Forest Machine Learning model to predict project budget based on project size, team size and duration."
)
'''

with open("app.py","w") as f:
    f.write(app_code)

# =========================
# Create requirements.txt
# =========================

requirements = '''
streamlit
pandas
scikit-learn
joblib
'''

with open("requirements.txt","w") as f:
    f.write(requirements)

# =========================
# Auto Download Files
# =========================

files.download("app.py")
files.download("requirements.txt")
files.download("model.pkl")
files.download("budget_dataset.csv")

print("All files created and downloaded successfully!")
```
